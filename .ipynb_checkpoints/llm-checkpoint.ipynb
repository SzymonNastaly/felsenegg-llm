````python
TARGET_VOCAB = 100

def list_of_unique_characters(text):
    res = set()
    for c in text:
        res.add(c)
    return sorted(list(res))
````

````python
vocab = list_of_unique_characters(text)
stoi = {token: integer_id for integer_id, token in enumerate(vocab)}
itos = {integer_id: token for integer_id, token in enumerate(vocab)}
number_merges_needed = TARGET_VOCAB - len(vocab)
for i in range(number_merges_needed):
    pair_counts = {}
    for i in range(len(vocab)-2):
        pair = (vocab[i], vocab[i+1])
        if not pair in pair_counts.keys():
            pair_counts[pair] = 0
        else:
            pair_counts[pair] += 1
    best_pair = max(pair_counts, key=pair_counts.get)
    vocab.append(best_pair)

    
````
